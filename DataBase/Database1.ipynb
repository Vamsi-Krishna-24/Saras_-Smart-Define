```python
def parse_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            parts = line.split('|')

            if len(parts)<2:
                continue

            words = parts[0].split()

            d=[]
            for word in words:
                if word.isalpha() and len(word)>1:
                    d.append(word)
                    print(d)    #Till here the first word is extracted as a list


            
            
                
            


```
