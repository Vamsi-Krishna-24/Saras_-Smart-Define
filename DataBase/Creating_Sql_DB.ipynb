```python
import sqlite3
```

```python
#creating one database, defining a FUNCTION

def create_database():
    connect = sqlite3.connect('Dictionary.db')
    cursor = connect.cursor()

    cursor.execute('''
            CREATE TABLE IF NOT EXISTS dictionary(
                   word1 TEXT,
                   synonm TEXT,
                   definition TEXT,
                   examples TEXT)
''')
    
    connect.commit()
    
```

```python
create_database()
```

```python
def insert_data(parsed_data):

    connect= sqlite3.connect('Dictionary.db')
    cursor = connect.cursor()

    for entry in parsed_data:
        word1=entry['word1']
        synonm=','.join(entry['synonm'])
        defintion = entry['definition']
        example = ','.join(entry['examples'])

        cursor.execute('''
        INSERT INTO dictionary (word1, synonm, definition, example)  
        VALUES(?,?,?,?)
        ''',(word1,synonm,defintion,example))

    connect.commit()
        
    
```

```python
insert_data()
```

```python
import sqlite3

# Function to connect and execute queries
def execute_query(query, fetch=False):
    connection = sqlite3.connect('Dictionary.db')
    cursor = connection.cursor()
    cursor.execute(query)
    
    if fetch:
        result = cursor.fetchall()
        connection.close()
        return result
    connection.commit()
    connection.close()

# Query to count all duplicate words, not just distinct ones
duplicates_query = '''
    SELECT word1, COUNT(word1) 
    FROM dictionary
    GROUP BY word1
    HAVING COUNT(word1) > 1
'''

# Execute the query and fetch results
duplicates = execute_query(duplicates_query, fetch=True)

# Display the total number of duplicate words (including all occurrences)
print(f"Duplicate words and their counts:")
for word, count in duplicates:
    print(f"'{word}': {count} occurrences")
```
